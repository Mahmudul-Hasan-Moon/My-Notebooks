### Decryption code

```python
# defination of XOR in two string
def xor(x, y):
    return '{1:0{0}b}'.format(len(x), int(x, 2) ^ int(y, 2))


f1 = open("codon.txt","r")      # Take the encrypted string from codon file

strr = f1.read()
ab = len(strr)   # length of encrypted message

print(ab)

aa = strr[ab-16:ab]    # for the length of the message
strr = strr[0:ab-16]    # remove the number of codon from string

ba = int(aa,2)       # size of codon 
print(ba-1)

print(aa)

print(strr)
#Number of codon in the encrypted string

key = bin(60)[2:]
print(key)
w = ba

k2 = 13                      # take a new key for creat the position where the codon are

R = int(800/8) % k2
x0 = 5

pos = []
pos.append(x0)
for i in range(w-1):  # create the position where codon is stored
    p = x0 + R
    x0 = p
    pos.append(p)

#print(R)
print(pos) 

substr = []

for i in range(w):       # extract the codon from the text    
    q = 0   
    j = 0
    q = pos[i]
    print(q)
    j = (q*2)-4
    k = (q*2)+2
    print(j,k)
    sub = strr[j:k]
    substr.append(sub)
    
    print(substr)

m = len(substr)

my_string = []

for i in range(m):             # XOP each codon with the first key
    r = xor(substr[i],key)
    my_string.append(r)
    
print(my_string)

s='0'
for i in range(m):       # after XOR, take all the codon with in a string
    s = s + my_string[i]
s = s[1:]
    
print(s)

binary = [s[i:i+8] for i in range(0, len(s), 8)]   # split the total string to many 8 character string to get ASCII
          
print(binary)

mn = len(binary)
message = '0'

for i in range(mn):    # convert each binary to ASCII and conver ASCII to Character and all character together make a string
    string = int(binary[i], 2)
    message = message + chr(string)
    print(string)  

message = message[1:]
print(message)
    
```
